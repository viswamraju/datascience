## Load necessary Libraries

github link:   https://github.com/KeithGalli/matplotlib_tutorial


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

Check documentation for more help

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.figure(figsize=(5, 3), dpi=200)    # changes the size of figure
# plt.plot(x, y, label="2x", color='r', linewidth=2, marker=".", markersize=10, markeredgecolor='b', linestyle='--')

# using fmt
# '[marker][line][color]'

plt.plot(x, y, '.-b', linewidth=2, label="2x")

plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})   # set title
plt.xlabel('X-axis', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})   # set xlabel
plt.ylabel('Y-axis')   # set y label

x2 = np.arange(0, 4.5, .5)
plt.plot(x2, x2**2, '*-r', label='square')

plt.xticks([0, 1, 2, 3])  # set x ticks
plt.yticks([2, 4, 6, 8])

plt.legend()
plt.savefig('mygraph.png')
plt.show()
```

### Bar Charts

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)
bars[0].set_hatch('O')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
```

## Real world Examples

### Line Graph


```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)
plt.show()
```

### show legends

```python

plt.plot(gas.Year, gas.USA, label="usa")
plt.plot(gas.Year, gas.Canada, label="canada")
plt.title('Gas prices over time in USD')    # set title
plt.legend()   # show legend
plt.show()
```

### Change the size of plot

```python
plt.figure(figsize=(10, 5))    # To chnage the size of figure
plt.plot(gas.Year, gas.USA, label="usa")
plt.plot(gas.Year, gas.Canada, label="canada")
plt.title('Gas prices over time in USD')    # set title
plt.legend()   # show legend
plt.show()
```

### Show x ticks

```python
plt.figure(figsize=(10, 5))    # To chnage the size of figure
plt.plot(gas.Year, gas.USA, '.-b', label="usa")
plt.plot(gas.Year, gas.Canada, '.-g', label="canada")
plt.plot(gas.Year, gas['South Korea'], '.-r', label='South Korea')
plt.title('Gas prices over time in USD')    # set title
plt.legend()   # show legend

plt.xlabel('Year')
plt.ylabel('Gas Price')

# plt.xticks(gas.Year)    # Set x ticks  it is clumsy if we have more data

plt.xticks(gas.Year[::3])   # Set x ticks using slicing   if we have more values we can use this

plt.show()
```

### Showing results for all countries

```python
plt.figure(figsize=(10, 5))    # To chnage the size of figure
# plt.plot(gas.Year, gas.USA, '.-b', label="usa")
# plt.plot(gas.Year, gas.Canada, '.-g', label="canada")
# plt.plot(gas.Year, gas['South Korea'], '.-r', label='South Korea')

for country in gas.loc[:, gas.columns != 'Year']:
    plt.plot(gas.Year, gas[country], marker=".", label=country)

plt.title('Gas prices over time in USD', fontdict={'fontweight': 'bold', 'fontsize': 20, 'fontname': 'Comic Sans MS'})    # set title
plt.legend()   # show legend

plt.xlabel('Year')
plt.ylabel('Gas Price')

# plt.xticks(gas.Year)    # Set x ticks  it is clumsy if we have more data

plt.xticks(gas.Year[::3].tolist() + [2011])   # Set x ticks using slicing   if we have more values we can use this.  If we want 2011 

plt.show()
```

## Load fifa data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

### Histograms

```python
bins = list(range(0,101,10))

plt.hist(fifa.Overall, bins=bins)
plt.ylabel('No of players')
plt.xlabel('Skill Level')
plt.title('Distribution of player skills in FIFA 2018')
plt.show()
```

## Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot']== 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0] 
```

```python
plt.pie([left, right])
plt.show()
```
